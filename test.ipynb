#### 머신러닝 기초
- 레이블(정답)이 있는 입력(input)데이터를 모델에 투입하면 머신러닝 모델이 입력 데이터와 접답 사이의 관계
- 컴퓨터가 스스로 데이터를 학습하여 문제를 해결하는 과정을 말한다.

### 지도 학습 vs 비지도 학습

1. 지도학습

    1.1.지도학습은 학습 과장에서 정답 Y가 주고, 입력 데이터 X와 출력 데이터 Y룰 모두 알고 있는 상태

    1.2.Y = aX + b 와 같이 X 와 Y사이의 관계를 알아내는 머신러닝 알고리즘이다.


2. 비지도 학습

    2.1. 정답Y가 주어지지 않고 X 데이터만 제공하는 머신러닝 유형이다.

    2.2. X 데이터 사이에 존재하는 패턴 또는 규칙을 찾는 것이다

### 회귀 vs 분류

1. 회귀

    1.1 변수(X)와 목표 변수(Y) 사이의 회귀 관계식을 갖는다.

2. 분류

    2.1 변수(X) 와 목표 변수(Y) 사이의 관계를 찾지만, 예측하려는 목표 레이블이 연속적이지 않고 0,1,2와 같이 이산적인 값을 갖는 경우를 말한다.
